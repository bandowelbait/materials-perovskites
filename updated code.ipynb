```python
import numpy as np
from pymatgen.io.cif import CifParser
from pymatgen.core.periodic_table import Element

mean_electron_affinity = []
max_electron_affinity = []
min_electron_affinity = []
std_electron_affinity = []

mean_ionization_energy_1 = []
max_ionization_energy_1 = []
min_ionization_energy_1 = []
std_ionization_energy_1 = []

mean_ionization_energy_2 = []
max_ionization_energy_2 = []
min_ionization_energy_2 = []
std_ionization_energy_2 = []

mean_ionization_energy_3 = []
max_ionization_energy_3 = []
min_ionization_energy_3 = []
std_ionization_energy_3 = []

# Loop through the CIF data
for r in cif_files:
    try:
        cif = r['cif']
        parser = CifParser.from_string(cif)
        structure = parser.get_structures()
        p_sulfide = structure[0]
        
        d1 = []  # List for electron affinity
        d2 = []  # List for 1st ionization energy
        d3 = []  # List for 2nd ionization energy
        d4 = []  # List for 3rd ionization energy

        for S in p_sulfide.species:
            element = Element(str(S))
            
            # Electron affinity
            d1.append(element.electron_affinity or 0)
            
            # Ionization energies (1st, 2nd, and 3rd)
            ionization_energies = element.ionization_energies
            d2.append(ionization_energies[0] if len(ionization_energies) > 0 else 0)  # 1st ionization energy
            d3.append(ionization_energies[1] if len(ionization_energies) > 1 else 0)  # 2nd ionization energy
            d4.append(ionization_energies[2] if len(ionization_energies) > 2 else 0)  # 3rd ionization energy
        
        # Replace None values with 0 in electron affinity and ionization energies
        d1 = [0 if v is None else v for v in d1]
        d2 = [0 if v is None else v for v in d2]
        d3 = [0 if v is None else v for v in d3]
        d4 = [0 if v is None else v for v in d4]

        # Calculate statistics for electron affinity
        mean_electron_affinity.append(np.mean(d1))
        max_electron_affinity.append(np.max(d1))
        min_electron_affinity.append(np.min(d1))
        std_electron_affinity.append(np.std(d1))

        # Calculate statistics for 1st ionization energy
        mean_ionization_energy_1.append(np.mean(d2))
        max_ionization_energy_1.append(np.max(d2))
        min_ionization_energy_1.append(np.min(d2))
        std_ionization_energy_1.append(np.std(d2))

        # Calculate statistics for 2nd ionization energy
        mean_ionization_energy_2.append(np.mean(d3))
        max_ionization_energy_2.append(np.max(d3))
        min_ionization_energy_2.append(np.min(d3))
        std_ionization_energy_2.append(np.std(d3))

        # Calculate statistics for 3rd ionization energy
        mean_ionization_energy_3.append(np.mean(d4))
        max_ionization_energy_3.append(np.max(d4))
        min_ionization_energy_3.append(np.min(d4))
        std_ionization_energy_3.append(np.std(d4))
    
    except Exception as e:
        print(f"Error processing structure: {e}")
        continue

# Now you have separate stats for 1st, 2nd, and 3rd ionization energies, 
# along with electron affinity, to be used as features.
```
